```python
labels=df['continent'].value_counts().keys()
certified=[]
denied=[]
for i in labels:
    con1=df['continent']==i
    con2=df['case_status']=='Certified'
    con3=df['case_status']=='Denied'
    cert_con=con1&con2
    denied_con=con1&con3
    certified.append(len(df[cert_con]))
    denied.append(len(df[denied_con]))
    d1=pd.DataFrame(zip(labels,certified,denied),
    columns=['Continent','Certified','Denied'])
    d1.set_index('Continent')
```

Numerical vs Numerical

$scatter-plot:$
- plt.scatter
- will take two arguments x-axis and y-axis
- Both variables should be numerical
- It provides relationship between two variables
    - Positvely related
     - Negatively related
     -  No relation

```python
x=[1,2,3,4,5]
y=[11,12,13,14,15]
# (1,11) (2,12) (3,13) (4,14) (5,15)
plt.scatter(x,y)
plt.plot(x,y)
```

```python
x=[1,2,3,4,5]
y=[11,11,11,11,11]
# (1,11) (2,12) (3,13) (4,14) (5,15)
plt.scatter(y,x)
plt.plot(y,x)
```

```python
x=[i for i in range(-4,5)]
y=[i*i for i in range(-4,5)]
plt.scatter(x,y)
plt.plot(x,y)
```

$Correlation-Coefiecient$
- Denoted with r
- Pearson correlation coefficient
- r varies from -1 to 1
- r= -1 to 0 : Negatively correlated
- r= 0 to 1: Postively Correlated
- r=0 : No relation
